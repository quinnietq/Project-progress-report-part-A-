```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Load the Dataset
file_path = "/content/weatherAUS.csv"  # Adjust the path if needed
dataset = pd.read_csv(file_path)

# Display dataset info
print("Dataset Overview:")
print(dataset.info())
print("\nFirst few rows of the dataset:")
print(dataset.head())

# Step 2: Data Cleaning and Preprocessing
# Handle missing values
num_cols = dataset.select_dtypes(include=['float64']).columns
dataset[num_cols] = dataset[num_cols].fillna(dataset[num_cols].median())

cat_cols = dataset.select_dtypes(include=['object']).columns
for col in cat_cols:
    dataset[col] = dataset[col].fillna(dataset[col].mode()[0])

# Convert categorical variables to numeric
dataset['RainToday'] = dataset['RainToday'].map({'Yes': 1, 'No': 0})
dataset['RainTomorrow'] = dataset['RainTomorrow'].map({'Yes': 1, 'No': 0})

# Convert Date to datetime and extract useful features
dataset['Date'] = pd.to_datetime(dataset['Date'], format='%d-%m-%Y')
dataset['Year'] = dataset['Date'].dt.year
dataset['Month'] = dataset['Date'].dt.month
dataset['Day'] = dataset['Date'].dt.day

# Drop unnecessary columns
dataset = dataset.drop(['Date'], axis=1)

# Step 3: Feature Engineering
# Calculate temperature range
dataset['TempRange'] = dataset['MaxTemp'] - dataset['MinTemp']

# Generate synthetic attenuation metrics for RF and FSO
def calculate_rf_attenuation(rainfall, humidity):
    return 0.05 * rainfall + 0.01 * humidity

def calculate_fso_attenuation(cloud, rainfall, humidity):
    return 0.1 * cloud + 0.07 * rainfall + 0.02 * humidity

dataset['RF_Attenuation'] = dataset.apply(
    lambda row: calculate_rf_attenuation(row['Rainfall'], row['Humidity3pm']), axis=1)
dataset['FSO_Attenuation'] = dataset.apply(
    lambda row: calculate_fso_attenuation(row['Cloud3pm'], row['Rainfall'], row['Humidity3pm']), axis=1)

# Display dataset after feature engineering
print("\nDataset after Feature Engineering:")
print(dataset[['Rainfall', 'Humidity3pm', 'Cloud3pm', 'RF_Attenuation', 'FSO_Attenuation']].head())

# Drop non-numeric columns for the correlation heatmap
numeric_data = dataset.select_dtypes(include=['float64', 'int64'])

# Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_data.corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Attenuation Analysis
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Rainfall', y='RF_Attenuation', label='RF Attenuation', color='blue', data=dataset)
sns.scatterplot(x='Rainfall', y='FSO_Attenuation', label='FSO Attenuation', color='red', data=dataset)
plt.title('RF and FSO Attenuation vs Rainfall')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Attenuation (dB/km)')
plt.legend()
plt.show()

# Step 5: Modeling RF/FSO Attenuation
# Define features and target variables for RF and FSO
features = ['Rainfall', 'Humidity3pm', 'Cloud3pm', 'TempRange']
target_rf = 'RF_Attenuation'
target_fso = 'FSO_Attenuation'

# Split data for RF Attenuation modeling
X_rf = dataset[features]
y_rf = dataset[target_rf]
X_rf_train, X_rf_test, y_rf_train, y_rf_test = train_test_split(X_rf, y_rf, test_size=0.2, random_state=42)

# Train a Random Forest Regressor for RF Attenuation
rf_model_rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model_rf.fit(X_rf_train, y_rf_train)
y_rf_pred = rf_model_rf.predict(X_rf_test)

# Evaluate RF model
print("\nRF Attenuation Model Performance:")
print("MSE:", mean_squared_error(y_rf_test, y_rf_pred))
print("R2 Score:", r2_score(y_rf_test, y_rf_pred))

# Split data for FSO Attenuation modeling
X_fso = dataset[features]
y_fso = dataset[target_fso]
X_fso_train, X_fso_test, y_fso_train, y_fso_test = train_test_split(X_fso, y_fso, test_size=0.2, random_state=42)

# Train a Random Forest Regressor for FSO Attenuation
rf_model_fso = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model_fso.fit(X_fso_train, y_fso_train)
y_fso_pred = rf_model_fso.predict(X_fso_test)

# Evaluate FSO model
print("\nFSO Attenuation Model Performance:")
print("MSE:", mean_squared_error(y_fso_test, y_fso_pred))
print("R2 Score:", r2_score(y_fso_test, y_fso_pred))
```

```python
# Define a hybrid communication decision function
def hybrid_decision(row):
    if row['Rainfall'] < 2 and row['Cloud3pm'] < 5:  # Clear weather conditions
        return "FSO"
    else:  # Rainy or cloudy conditions
        return "RF"

# Apply the decision function to the dataset
dataset['Communication_Mode'] = dataset.apply(hybrid_decision, axis=1)

# Count the number of RF and FSO decisions
print("Communication Mode Distribution:")
print(dataset['Communication_Mode'].value_counts())

# Visualize the decision distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='Communication_Mode', data=dataset, palette='viridis')
plt.title('Distribution of Hybrid Communication Decisions')
plt.xlabel('Communication Mode')
plt.ylabel('Count')
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingRegressor

# Train Gradient Boosting for RF Attenuation
gbr_model_rf = GradientBoostingRegressor(n_estimators=100, random_state=42)
gbr_model_rf.fit(X_rf_train, y_rf_train)
y_rf_pred_gbr = gbr_model_rf.predict(X_rf_test)

# Evaluate Gradient Boosting for RF Attenuation
print("\nGradient Boosting RF Attenuation Model Performance:")
print("MSE:", mean_squared_error(y_rf_test, y_rf_pred_gbr))
print("R2 Score:", r2_score(y_rf_test, y_rf_pred_gbr))

# Train Gradient Boosting for FSO Attenuation
gbr_model_fso = GradientBoostingRegressor(n_estimators=100, random_state=42)
gbr_model_fso.fit(X_fso_train, y_fso_train)
y_fso_pred_gbr = gbr_model_fso.predict(X_fso_test)

# Evaluate Gradient Boosting for FSO Attenuation
print("\nGradient Boosting FSO Attenuation Model Performance:")
print("MSE:", mean_squared_error(y_fso_test, y_fso_pred_gbr))
print("R2 Score:", r2_score(y_fso_test, y_fso_pred_gbr))
```

```python
# Random Forest Feature Importance for RF Attenuation
rf_feature_importance = pd.Series(rf_model_rf.feature_importances_, index=features)
rf_feature_importance.sort_values(ascending=False, inplace=True)

# Visualize feature importance for RF
plt.figure(figsize=(8, 5))
rf_feature_importance.plot(kind='bar', color='skyblue')
plt.title('Feature Importance for RF Attenuation')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.show()

# Random Forest Feature Importance for FSO Attenuation
fso_feature_importance = pd.Series(rf_model_fso.feature_importances_, index=features)
fso_feature_importance.sort_values(ascending=False, inplace=True)

# Visualize feature importance for FSO
plt.figure(figsize=(8, 5))
fso_feature_importance.plot(kind='bar', color='orange')
plt.title('Feature Importance for FSO Attenuation')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.show()
```

```python
# Scatter plot with decision overlay
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Rainfall', y='Cloud3pm', hue='Communication_Mode', data=dataset, palette='viridis')
plt.title('Hybrid Communication Decision Based on Weather Conditions')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Cloud Cover (3 PM)')
plt.legend(title='Communication Mode')
plt.show()
```
